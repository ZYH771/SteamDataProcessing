```python
import numpy as np
import pandas as pd
import json
```

```python
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from nltk.stem import WordNetLemmatizer
```

```python
lemma = WordNetLemmatizer()
```

```python
from nltk.corpus import stopwords
import string
extras = ["☐", "``", "\n", "--", "☑'", "''", "game"]
stop = set(stopwords.words('english') + list(string.punctuation) + extras)
```

```python
file = open('reviews-list-1k (ind 12841).json')
review_dict = json.load(file)
```

```python
review_dict["1119730"][0]['review']
```

```python
import re
def checkString(str):
    match = re.search(r'[a-zA-Z]+', str)
    
    result = 0
    for char in str:
        result += 1
    if len(str) != 0:    
        letter_percent = result / len(str)
    else:
        letter_percent = 0
        
    if match and letter_percent > 0.7:
        return True
    else:
        return False
```

```python
def lengthCheck(str):
    if len(str) < 20:
        return False
    else:
        return True
```

```python
from itertools import islice

def dict_head(mydict):
    head = dict(islice(mydict.items(), 0, 2))
    return head
```

```python
for key in review_dict:
    limit = 999
    for j in range(len(review_dict[key])):
        if j > limit:
            break
        review = review_dict[key][j]['review']
        letter_chk = checkString(review)
        if letter_chk:
            review_dict[key][j].clear()
            review_token = word_tokenize(review.lower())
            review_clean_token = [lemma.lemmatize(i) for i in review_token if i not in stop]
            review_dict[key][j]['review'] = review_clean_token
        else:
            del review_dict[key][j]
            limit -= 1
```

```python
ids = review_dict.keys()
for id in ids:
    review_tokens = review_dict[id]
    all_tokens = []
    temp_list = []
    for review in review_tokens:
        temp_list = review['review']
        for i in range(len(temp_list)):
            if len(temp_list[i]) > 3:
                all_tokens.append(temp_list[i])
        temp_list = []
    review_dict[id] = all_tokens
```

```python
for id in ids:
    common_list = []
    review_tokens = review_dict[id]
    fdist = FreqDist(review_tokens)
    common = fdist.most_common(10)
    for w in common:
        common_list.append(w[0])
    review_dict[id] = common_list
```
